```python
import pandas as pd
```

```python
cluster = pd.read_csv('clustered.csv')
g_cluster = cluster[['lat','lng']]
g_cluster = g_cluster.append({'lng':126.842948,'lat':35.226879},ignore_index=True)
```

```python
from api import get_api
import requests
import json
```

```python
def calc_distance(originX,originY,destinationX, destinationY):
    # origin and destination should be coordinate
    url = 'https://apis-navi.kakaomobility.com/v1/directions'
    params = {'origin': str(originX) +','+str(originY),
              'destination': str(destinationX) + ','+str(destinationY)}
    header = {'Authorization': 'KakaoAK ' + get_api()}

    route = requests.get(url, params=params, headers=header).json()
#     print(route)

    if route['routes'][0]['result_code'] == 104:
        return 0
    else:
        return(route['routes'][0]['summary']['duration'])
```

```python
# distance_matrix 계산하는 부분

sample = [0]*len(g_cluster)
distance_matrix = []

for i in range (len(g_cluster)):
    single_distance_matrix = []
    if i == 35:
        distance_matrix.append(sample)
        continue
        
    for j in range(len(g_cluster)):
#         print(g_cluster.iloc[i]['x'])
        distance = calc_distance(g_cluster.iloc[i]['lng'],g_cluster.iloc[i]['lat'],g_cluster.iloc[j]['lng'],g_cluster.iloc[j]['lat'])
        single_distance_matrix.append(distance)
        if(j == 20):
            print("phase" + str(i))
    print(single_distance_matrix)
    distance_matrix.append(single_distance_matrix)
    
distance_matrix
```

```python
distance_dataframe = pd.DataFrame(distance_matrix)
distance_dataframe.to_csv('distance.csv',mode='w',index=False)
```
