```python
import pandas as pd
```

```python
cluster = pd.read_csv('cluster_info.csv')
```

```python
import folium
cluster
```

```python
m = folium.Map( location = [35.212616,126.842246],zoom_start=13)
```

```python
for i in range(len(cluster)):
    folium.Marker([cluster.iloc[i]['lat'],cluster.iloc[i]['lng']]).add_to(m)
```
