```python
import json
import requests
```

```python
# 위도 경도 가져오기
def convert_location(region):

    url = 'https://dapi.kakao.com/v2/local/search/address'
    params = {'query': region, 'page': 1}
    header = {'Authorization': 'KakaoAK ' + get_api()}

    places = requests.get(url, params=params, headers=header).json()[
        'documents']

    return {'x': places[0]['x'], 'y': places[0]["y"]}
```

```python
x = []
y = []

for add in df['address']:
#     print(add)
    data = convert_location(add)
    x.append(data['x'])
    y.append(data['y'])
    
```

```python
df['x']=x
df['y']=y
```

```python
def calc_distance(originX,originY,destinationX, destinationY):
    # origin and destination should be coordinate
    url = 'https://apis-navi.kakaomobility.com/v1/directions'
    params = {'origin': originX +','+originY,
              'destination': destinationX + ','+destinationY}
    header = {'Authorization': 'KakaoAK ' + get_api()}

    route = requests.get(url, params=params, headers=header).json()
#     print(route['routes'][0]['result_code'])
    if route['routes'][0]['result_code'] == 104:
        return 0
    else:
        return(route['routes'][0]['summary']['duration'])
```
