```python
# 위도 경도 가져오기
def convert_location(region):

    url = 'https://dapi.kakao.com/v2/local/search/address'
    params = {'query': region, 'page': 1}
    header = {'Authorization': 'KakaoAK ' + get_api()}

    places = requests.get(url, params=params, headers=header).json()[
        'documents']

    return {'x': places[0]['x'], 'y': places[0]["y"]}
```

```python
def create_data_model():
    data = {}
    data['distance_matrix'] = relist
    data['num_vehicles'] = 5
    data['depot'] = 0
    return data
```

```python
import matplotlib.pyplot as plt
from ortools.constraint_solver import routing_enums_pb2
from ortools.constraint_solver import pywrapcp
```

```python
# Instance the data problem
data = create_data_model()
```

```python
# Create the routing index manager
manager = pywrapcp.RoutingIndexManager(len(data['distance_matrix']),data['num_vehicles'], data['depot'])
```

```python
# Create Routing Model.
routing = pywrapcp.RoutingModel(manager)
```

```python
# Create and register a transit callback.
def distacne_callback(from_index, to_index):
    """Returns the travel time between the two nodes."""
    # Convert from routing variable Index to time matrix NodeIndex.
    from_node = manager.IndexToNode(from_index)
    to_node = manager.IndexToNode(to_index)
    return data['distance_matrix'][from_node][to_node] + data['service_time'][from_node]
```

```python
transit_callback_index = routing.RegisterTransitCallback(distacne_callback)
```

```python
# Define cost of each arc.
routing.SetArcCostEvaluatorOfAllVehicles(transit_callback_index)# Define cost of each arc.
```

```python
# Define cost of each arc.
routing.SetArcCostEvaluatorOfAllVehicles(transit_callback_index)

# Add Distance constraint.
dimension_name = 'Distance'
routing.AddDimension(
    transit_callback_index,
    0,  # no slack
    3600,  # vehicle maximum travel distance
    True,  # start cumul to zero
    dimension_name)
distance_dimension = routing.GetDimensionOrDie(dimension_name)
distance_dimension.SetGlobalSpanCostCoefficient(100)
```

```python
# Setting first solution heuristic
search_parameters = pywrapcp.DefaultRoutingSearchParameters()
search_parameters.first_solution_strategy = (
    routing_enums_pb2.FirstSolutionStrategy.PATH_CHEAPEST_ARC)
```

```python
solution = routing.SolveWithParameters(search_parameters)
```

```python
def print_solution(data, manager, routing, solution):
    """Prints solution on console."""
    print(f'Objective: {solution.ObjectiveValue()}')
    time_dimension = routing.GetDimensionOrDie('Time')
    total_time = 0
    for vehicle_id in range(data['num_vehicles']):
        index = routing.Start(vehicle_id)
        plan_output = 'Route for vehicle {}:\n'.format(vehicle_id)
        while not routing.IsEnd(index):
            time_var = time_dimension.CumulVar(index)
            plan_output += '{0} Time({1},{2}) -> '.format(
                manager.IndexToNode(index), solution.Min(time_var),
                solution.Max(time_var))
            index = solution.Value(routing.NextVar(index))
        time_var = time_dimension.CumulVar(index)
        plan_output += '{0} Time({1},{2})\n'.format(manager.IndexToNode(index),
                                                    solution.Min(time_var),
                                                    solution.Max(time_var))
        plan_output += 'Time of the route: {}min\n'.format(
            solution.Min(time_var))
        print(plan_output)
        total_time += solution.Min(time_var)
    print('Total time of all routes: {}min'.format(total_time))
```

```python
# Print solution on console.
if not solution:
    print("There is no solution.")
if solution:
    print_solution(data, manager, routing, solution)
```
